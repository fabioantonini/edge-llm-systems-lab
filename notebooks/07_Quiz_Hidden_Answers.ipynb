```python
import os
import requests

# Resolve Ollama endpoint depending on environment
BASE_URL = os.getenv('OLLAMA_BASE_URL', 'http://localhost:11434')
print('Using Ollama at:', BASE_URL)

# Diagnostic check
try:
    r = requests.get(BASE_URL, timeout=3)
    print('Server reachable. Status:', r.status_code)
    try:
        tags = requests.get(f"{BASE_URL}/api/tags", timeout=3)
        print('Tags endpoint status:', tags.status_code)
    except Exception as e:
        print('Tags endpoint check failed:', e)
except Exception as e:
    print('Server not reachable:', e)
```

## Question 1
**What problem does Ollama solve compared to using cloud LLM APIs?**

<details>
<summary><b>Answer</b> (click to reveal)</summary>

It provides a simple way to run open-source LLMs locally (offline, no API keys, no per-token cost) while exposing a consistent CLI and REST API for integration.

</details>

## Question 2
**Which default port does the Ollama server use?**

<details>
<summary><b>Answer</b> (click to reveal)</summary>

Port **11434**.

</details>

## Question 3
**What is the difference between `/api/generate` and `/api/chat` in Ollama?**

<details>
<summary><b>Answer</b> (click to reveal)</summary>

`/api/generate` is typically single-turn text generation; `/api/chat` supports multi-turn messages with roles (user/assistant/system).

</details>

## Question 4
**Why are quantized models useful for CPU inference?**

<details>
<summary><b>Answer</b> (click to reveal)</summary>

Quantization reduces precision (e.g., 4-bit/8-bit), which reduces memory footprint and can increase speed, making inference feasible on CPUs.

</details>

## Question 5
**What is “local inference latency” and why should students measure it?**

<details>
<summary><b>Answer</b> (click to reveal)</summary>

Latency is the time from prompt to output. Measuring it highlights model-size and hardware trade-offs and makes performance constraints tangible.

</details>

## Question 6
**Name two CPU-friendly model families often used with Ollama.**

<details>
<summary><b>Answer</b> (click to reveal)</summary>

Examples: **Llama (small variants like 3B)**, **Phi**, **Gemma (2B)**, **Qwen (3B)** (any two).

</details>

## Question 7
**What is the minimum information you typically send in a request to `/api/generate`?**

<details>
<summary><b>Answer</b> (click to reveal)</summary>

A JSON payload with at least `model` and `prompt` (and often `stream: false`).

</details>

## Question 8
**What does `temperature` control in text generation?**

<details>
<summary><b>Answer</b> (click to reveal)</summary>

It controls randomness/creativity: higher temperature increases variability; lower temperature makes outputs more deterministic.

</details>

## Question 9
**What is RAG (Retrieval-Augmented Generation) in one sentence?**

<details>
<summary><b>Answer</b> (click to reveal)</summary>

RAG retrieves relevant documents (context) and uses an LLM to generate an answer grounded in that retrieved context.

</details>

## Question 10
**Why use a vector database in RAG?**

<details>
<summary><b>Answer</b> (click to reveal)</summary>

To store embeddings and perform fast similarity search to find the most relevant chunks for a query.

</details>

## Question 11
**What is an embedding (in the RAG context)?**

<details>
<summary><b>Answer</b> (click to reveal)</summary>

A vector representation of text in a semantic space where similar meanings are close.

</details>

## Question 12
**What can go wrong if you skip chunking and embed huge documents as a single vector?**

<details>
<summary><b>Answer</b> (click to reveal)</summary>

Retrieval becomes coarse: you may retrieve irrelevant content, lose granularity, and reduce answer quality.

</details>

## Question 13
**Why is it important to instruct the LLM to answer using ONLY the provided context in RAG demos?**

<details>
<summary><b>Answer</b> (click to reveal)</summary>

To demonstrate grounding and reduce hallucinations; it also makes evaluation clearer.

</details>

## Question 14
**In a Docker Compose setup, why might the notebook use `http://ollama:11434` instead of `http://localhost:11434`?**

<details>
<summary><b>Answer</b> (click to reveal)</summary>

Inside Docker networking, `localhost` refers to the notebook container. `ollama` is the service hostname on the Compose network.

</details>

## Question 15
**Give one reason to prefer local LLMs in an industrial/engineering setting.**

<details>
<summary><b>Answer</b> (click to reveal)</summary>

Data privacy (no data leaves the machine), offline operation, predictable cost, or reproducibility.

</details>

## Question 16
**What is the main limitation of running larger LLMs on CPU?**

<details>
<summary><b>Answer</b> (click to reveal)</summary>

Higher latency and memory usage; may become too slow or exceed RAM.

</details>

## Question 17
**How can you observe CPU/RAM usage during inference?**

<details>
<summary><b>Answer</b> (click to reveal)</summary>

Use Task Manager on Windows, `top/htop` on Linux, or `psutil` from Python.

</details>

## Question 18
**What is a simple baseline retrieval method that does NOT require neural embeddings?**

<details>
<summary><b>Answer</b> (click to reveal)</summary>

TF-IDF / bag-of-words similarity (e.g., scikit-learn `TfidfVectorizer`).

</details>

## Question 19
**Why might outputs differ across runs even with the same prompt?**

<details>
<summary><b>Answer</b> (click to reveal)</summary>

Stochastic sampling (temperature/top-p), different seeds, or different model versions/settings.

</details>

## Question 20
**What is a safe way to teach prompt engineering in this lab?**

<details>
<summary><b>Answer</b> (click to reveal)</summary>

Have students compare structured vs unstructured prompts, add constraints (format, length), and evaluate consistency and groundedness.

</details>
